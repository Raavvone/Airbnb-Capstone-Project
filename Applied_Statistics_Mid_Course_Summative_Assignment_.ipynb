```python
# Calculate measures of central tendency (mean, median, mode)
mean_values = df.mean()
median_values = df.median()
mode_values = df.mode().iloc[0]

# Display measures of central tendency
central_tendency = pd.DataFrame({
    'Mean': mean_values,
    'Median': median_values,
    'Mode': mode_values
})
print("Measures of Central Tendency:")
print(central_tendency)
```

```python
# Calculate measures of dispersion (variance, standard deviation)
variance_values = df.var()
std_deviation_values = df.std()

# Display measures of dispersion
dispersion = pd.DataFrame({
    'Variance': variance_values,
    'Standard Deviation': std_deviation_values
})
print("\nMeasures of Dispersion:")
print(dispersion)
```

```python
from scipy.stats import binom

# Determine parameters for the Binomial distribution
n = df['NumDealsPurchases'].max()  # Maximum number of trials
p = df['NumDealsPurchases'].mean() / n  # Probability of success (mean divided by maximum)

# Calculate probabilities based on the Binomial distribution
binom_probabilities = binom.pmf(k=df['NumDealsPurchases'], n=n, p=p)

# Calculate expected value based on the Binomial distribution
binom_expected_value = binom.mean(n=n, p=p)

# Display results for the Binomial distribution
print("Binomial Distribution:")
print("Probabilities:", binom_probabilities)
print("Expected Value:", binom_expected_value)
```

```python
from scipy.stats import norm

# Determine parameters for the Normal distribution
normal_mean = df['Income'].mean()
normal_std_dev = df['Income'].std()

# Calculate probabilities based on the Normal distribution
norm_probabilities = norm.pdf(x=df['Income'], loc=normal_mean, scale=normal_std_dev)

# Calculate expected value based on the Normal distribution
norm_expected_value = norm.mean(loc=normal_mean, scale=normal_std_dev)

# Display results for the Normal distribution
print("\nNormal Distribution:")
print("Probabilities:", norm_probabilities)
print("Expected Value:", norm_expected_value)
```

```python
# Explore relationships between customer characteristics and spending habits
# For example, you can use correlation analysis or visualize relationships using scatter plots, heatmaps, or pair plots
# Example:
import seaborn as sns

# Pair plot to visualize relationships between numerical variables
sns.pairplot(df[['Year_Birth', 'Income', 'MntFishProducts', 'MntMeatProducts', 'MntFruits', 'MntWines']])
plt.show()

# Heatmap to visualize correlation matrix between numerical variables
corr_matrix = df[['Year_Birth', 'Income', 'MntFishProducts', 'MntMeatProducts', 'MntFruits', 'MntWines']].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```

```python
from sklearn.cluster import KMeans
import numpy as np

# Select relevant features for segmentation
X = df[['Year_Birth', 'Income', 'MntFishProducts', 'MntMeatProducts', 'MntFruits', 'MntWines']]

# Scale the features
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Determine the optimal number of clusters using the elbow method
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()

# Apply K-means clustering with the optimal number of clusters
k = 3  # Example: Assuming 3 clusters
kmeans = KMeans(n_clusters=k, init='k-means++', random_state=42)
kmeans.fit(X_scaled)

# Add cluster labels to the dataframe
df['Cluster'] = kmeans.labels_

# Visualize the clusters
sns.scatterplot(data=df, x='Income', y='MntWines', hue='Cluster', palette='viridis')
plt.title('Customer Segmentation')
plt.xlabel('Income')
plt.ylabel('Amount spent on Wines')
plt.show()
```

```python
# Analyze key insights from customer segmentation
# For example, you can examine the characteristics and spending habits of each customer segment
# Example:
segment_means = df.groupby('Cluster').mean()
print(segment_means)
```

```python
# Visualize key insights and findings
# Example visualizations:
import matplotlib.pyplot as plt

# Example: Bar plot to visualize average spending on different product categories by cluster
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='Cluster', y='MntFishProducts', ci=None)
plt.title('Average Spending on Fish Products by Cluster')
plt.xlabel('Cluster')
plt.ylabel('Average Spending on Fish Products')
plt.show()

# Example: Box plot to visualize distribution of income by cluster
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Cluster', y='Income')
plt.title('Distribution of Income by Cluster')
plt.xlabel('Cluster')
plt.ylabel('Income')
plt.show()
```
