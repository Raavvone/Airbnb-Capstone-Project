```python
df2 = df.loc[df["availability_365"] > 0 ]

#box plot for availability room
plt.title("Neighbourhood Group vs. Availability Room")
sns.boxplot(data=df2, x='neighbourhood_group',y='availability_365',palette="dark")
plt.show()
```

Insight-From this analysis we came to know that eventhough Broklyn and Manhattan has higest number of rooms but there availabilty is very lesss when compared to staten Island. This can be one of the reason where customer wil miiss the room in this region because of there lees availability.

```python
df3 = df.loc[df["price"] > 0 ]
# boxplot for price distribution
plt.style.use('classic')
plt.figure(figsize=(13,7))
plt.title("Neighbourhood Group Price Distribution < 500")
sns.boxplot(y="price",x ='neighbourhood_group' ,data = df3[df3.price<500])
plt.show()
```

Insight- this shows the price distribution in different neighbourhood groups where Manhattan is much costlier followed by Brooklyn and Queens

```python
corr = df.corr()
plt.figure(figsize=(13,10))
plt.title("Correlation Between Different Variables\n")
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.show()
```
